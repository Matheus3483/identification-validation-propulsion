```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.style.use('default')
```

```python
def number_of_zeros(x):
    for i in range(x.size):
        if x[i] != 0:
            return i
    return 0
```

```python
df1 = pd.read_csv('../data/raw/cossenos_n10_emax_tempo_9v.csv', header=None)
data_array = np.array(df1)
delta_tempo_array = (data_array[:,0] << 16) + (data_array[:,1] << 8) + data_array[:,2]
qnt_zeros = number_of_zeros(delta_tempo_array)
freq_array_cossenos_emax_tempo_9v = np.zeros(delta_tempo_array.shape)
freq_array_cossenos_emax_tempo_9v[qnt_zeros:] = 1/7/(delta_tempo_array[qnt_zeros:]*1e-5) * (2*np.pi) # removido os zeros de "delta_tempo_array" 
entrada_array_cossenos_emax_tempo_9v = (data_array[:,3]-125)/(249-125) * 9
df2 = pd.DataFrame([entrada_array_cossenos_emax_tempo_9v,freq_array_cossenos_emax_tempo_9v]).T
df2.to_csv('../data/entrada_saida_cossenos_n10_emax_tempo_9v.csv',header=False,index=False)
```
